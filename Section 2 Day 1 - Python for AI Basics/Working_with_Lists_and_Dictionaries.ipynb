# Working with Lists and Dictionaries

```python
# Lists

nuumbers = [1, 2, 3, 4 ,5]
print(nuumbers[3])

nuumbers.append(6)
print(nuumbers)
```

```python
# Dictionaries

student = {"name": "Scott", "age": 40, "is_student": True}
print(student["age"])
print(student["name"])
print(student["is_student"])
```
