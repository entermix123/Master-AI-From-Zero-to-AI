# Introduction to NumPy

```python
# Creating Array

import numpy as np

array = np.array([1,2,3,4,5])
print(array)
```

```python
array_2 = np.array([5,4,3,2,1])
print(array + array_2)
```

```python
matrix = np.array([[1,2],
        		   [3,4]])
# print matrix
print(matrix)

# matrices multplication
print(np.dot(matrix, matrix))
```
