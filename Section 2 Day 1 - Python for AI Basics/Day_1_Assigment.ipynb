```python
# Exercise 3: Working with Lists and Dictionaries
# Modify the above code to read a CSV file, select specific columns, and write the output to a new CSV file.

# Step 1: Create a list of students
students = ["Alice", "Bob", "Charlie", "David", "Eve"]

# Step 2: Create a dictionary with their ages
ages = {
    "Alice": 18,
    "Bob": 25,
    "Charlie": 19,
    "David": 16,
    "Eve": 22
}

# Step 3: Iterate through the dictionary and print each student's name and age
for student in students:
    print(f"{student} is {ages[student]} years old")

# Optional Step 4: Write to a CSV file
import csv

with open("students_output.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Age"])  # headers
    for student in students:
        writer.writerow([student, ages[student]])

print("students_output.csv created successfully!")
```
