# Titanic exercise

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load all three parts
df1 = pd.read_csv('titanic_1.csv')
df2 = pd.read_csv('titanic_2.csv')
df3 = pd.read_csv('titanic_3.csv')

# Combine them into one dataset
data = pd.concat([df1, df2, df3], ignore_index=True)

# Fill missing data
data.fillna({'Age': data['Age'].mean()}, inplace=True)

# Analyze suvival rate
print(data['Survived'].value_counts(normalize=True))

# Analyze Survival by sex
print(data.groupby(['Sex', 'Pclass'])['Survived'].mean())

# Visualize the data
sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Survival Rate by Gender')
plt.show()
```
