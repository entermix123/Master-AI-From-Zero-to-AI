# Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all three parts
df1 = pd.read_csv('titanic_1.csv')
df2 = pd.read_csv('titanic_2.csv')
df3 = pd.read_csv('titanic_3.csv')

# Combine into one dataset
data = pd.concat([df1, df2, df3], ignore_index=True)

# Count survival values
survival_counts = data['Survived'].value_counts()

# Create labels
labels = ['Did Not Survive', 'Survived']

# Create pie chart
plt.figure()
plt.pie(
    survival_counts,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Proportion of Passengers Who Survived vs Did Not Survive')
plt.show()
```
