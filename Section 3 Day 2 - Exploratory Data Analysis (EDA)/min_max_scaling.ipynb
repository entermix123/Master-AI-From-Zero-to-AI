# Min Max Scaling

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

data = pd.DataFrame({
    "Age": [18, 25, 19, 16, 22],
    "Income": [2000, 3500, 2500, 1800, 3000]
})

scaler = MinMaxScaler()

data[['Age', 'Income']] = scaler.fit_transform(data[['Age', 'Income']])

print(data)
```

```python
import pandas as pd

# Sample dataset
data = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Gender": ["Female", "Male", "Male", "Male", "Female"],
    "Age": [18, 25, 19, 16, 22],
    "Purchased": ["Yes", "No", "Yes", "No", "Yes"]
})

print("Original Data:\n")
print(data)

# Apply one-hot encoding to Gender and Purchased
data_encoded = pd.get_dummies(data, columns=['Gender', 'Purchased'])

print("\nOne-Hot Encoded Data:\n")
print(data_encoded)
```
